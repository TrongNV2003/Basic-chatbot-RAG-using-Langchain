```python
model_path = "Trongdz/Llama-1.1B-AWQ-4bit"
#load the model and tokenizer

model = AutoAWQForCausalLM.from_pretrained(model_path)
tokenizer = AutoTokenizer.from_pretrained(model_path, use_fast=True)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
```

```python
prompt = "What planets are in our solar system?"

formatted_prompt = f"<|im_start|>user\n{prompt}<|im_end|>\n<|im_start|>assistant"

tokens = tokenizer(
    formatted_prompt,
    return_tensors='pt'
).input_ids.cuda()

generation_output = model.generate(
    tokens,
    do_sample = False,
    max_new_tokens = 512
)

output = tokenizer.decode(generation_output[0], skip_special_token=True)
print(output)
```
